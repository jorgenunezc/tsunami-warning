```python
import os
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = os.getcwd()
folder= path + "/dataset/series_boyas/analisis/"
print(folder)
```

```python
df = pd.read_json(folder+'DataFrame_boyas.json') # cargar dataframe
```

```python
df.head()
```

```python
df.describe()
```

```python
print(df.isna().sum())
```

```python
columns = df.columns
```

```python
files = []
for file in columns:
    if file.endswith("E3_9.0"):
        files.append(file)
files.sort()
print(files)
```

```python
plt.figure(figsize=(24,24))
sns.heatmap(df[files].corr(), vmax=1.0,vmin = -1.0 ,linewidths=0.1,
            square=True,annot=True,cmap='viridis',linecolor="k")
sns.set(font_scale=2)
plt.title('Correlacion entre caracteristicas')
plt.rc('xtick', labelsize=17) 
plt.rc('ytick', labelsize=17)
plt.xticks(rotation='vertical')
plt.yticks(rotation='horizontal')
plt.rcParams["font.weight"] = "bold"
plt.show()
```
